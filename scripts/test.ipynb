```python
from pathlib import Path
import re

dir_path = Path("/data/raw_data/mimic/files/mimiciv/meds_v0.6/3.1/etho_meds/tuning")  # TODO: change me
offset = 292

# match files like "0.parquet", "1.parquet", ...
pat = re.compile(r"^(\d+)\.parquet$")

# Sort by numeric id for determinism
files = []
for p in dir_path.iterdir():
    m = pat.match(p.name)
    if m:
        files.append((int(m.group(1)), p))
files.sort(key=lambda x: x[0])

for i, p in files:
    new_name = f"{i + offset}.parquet"
    new_path = p.with_name(new_name)
    if new_path.exists():
        raise FileExistsError(f"Target already exists: {new_path}")
    p.rename(new_path)

print(f"Renamed {len(files)} files.")
```
